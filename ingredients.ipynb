```python
new = pd.DataFrame(ingredients, columns=['ingredients'])
new.head()
```

```python
new['ingredients'] = new['ingredients'].apply(lambda x: x.lower())
new['ingredients'] = new['ingredients'].apply(lambda x: x.strip())
```

```python
new.head()
```

```python
def find_word(word):
    """
    Displays the list of words containing the given letters
    
    Parameters
    ----------    
    word : str
        Letters to be used
        'tom'
    """
    filt= new['ingredients'].str.contains(word, na=False)
    print(new.loc[filt,'ingredients'])
    
def change_word(list_words, new_word, word):
    """
    Replace the words given in the list by the new word
    and call the function find_word
    
    Parameters
    ----------
    list_words : list
        Word to be replaced
        ['tomatoes', 'tomato paste', 'tomato sauce', 'chopped tomatoes']
    new_word : str
        Word to be used
        'tomato'
    word : str
        Letters to be used
        'tom'
    """
    
    pattern = '|'.join(list_words)
    new['ingredients'] = new['ingredients'].str.replace(pattern, new_word, regex=True)
    find_word(word)
    #filt= new['ingredients'].str.contains('jag', na=False)
    #new.loc[filt,'ingredients']
```

```python
ingredients = new.pivot_table(index = ['ingredients'], aggfunc ='size')
ingredients = ingredients.to_frame()
ingredients.reset_index(inplace=True)
ingredients = ingredients.rename(columns = {0:'n_ingredients', 'ingr':'ingredient'})
ingredients
```

```python
most_used = ingredients.nlargest(15, ['n_ingredients'])
most_used
```

```python
plt.figure()
sns.catplot(x='n_ingredients', y='ingredients', kind='bar', data=most_used)
plt.title('Most common ingredients in Indian food')
plt.xlabel('number of times')
plt.ylabel('ingredients')
plt.show()
```

```python
filt= (ingredients['ingredients'].str.contains('flour', na=False)) | \
      (ingredients['ingredients'].str.contains('semolina', na=False))
flour = ingredients.loc[filt]
flour = flour.sort_values('n_ingredients', ascending=False)
```

```python
plt.figure()
sns.catplot(x='n_ingredients', y='ingredients', kind='bar', data=flour)
plt.title('Flour', x=-0.35, y=1.10, fontsize=18)
plt.suptitle('Different types of flour used in Indian cooking', x=0.3, y=1.05, fontsize=10)
plt.xlabel('number of times')
plt.ylabel('Type of flour')
plt.show()
```

```python
filt= (ingredients['ingredients'].str.contains('oil', na=False))
oil = ingredients.loc[filt]
oil = oil.sort_values('n_ingredients', ascending=False)
```

```python
plt.figure()
sns.catplot(x='n_ingredients', y='ingredients', kind='bar', data=oil)
plt.title('Types of oil', x=0.35, y=1.10, fontsize=18)
plt.suptitle('Different types of flour used in Indian cooking', y=1.05, fontsize=10)
plt.xlabel('Number of times')
plt.ylabel('Type of oil')
plt.show()
```
