```python
def find_word(word):
    """
    Displays the list of words containing the given letters
    
    Parameters
    ----------    
    word : str
        Letters to be used
        'tom'
    """
    filt= new['ingredients'].str.contains(word, na=False)
    print(new.loc[filt,'ingredients'])
    
def change_word(list_words, new_word, word):
    """
    Replace the words given in the list by the new word
    and call the function find_word
    
    Parameters
    ----------
    list_words : list
        Word to be replaced
        ['tomatoes', 'tomato paste', 'tomato sauce', 'chopped tomatoes']
    new_word : str
        Word to be used
        'tomato'
    word : str
        Letters to be used
        'tom'
    """
    
    pattern = '|'.join(list_words)
    new['ingredients'] = new['ingredients'].str.replace(pattern, new_word, regex=True)
    find_word(word)
    #filt= new['ingredients'].str.contains('jag', na=False)
    #new.loc[filt,'ingredients']
```

```python
ingredients = new.pivot_table(index = ['ingredients'], aggfunc ='size')
ingredients = ingredients.to_frame()
ingredients.reset_index(inplace=True)
ingredients = ingredients.rename(columns = {0:'n_ingredients', 'ingr':'ingredient'})
ingredients
```

```python
most_used = ingredients.nlargest(15, ['n_ingredients'])
most_used
```

```python
import matplotlib
plt.figure(figsize=(20,1))

plot = sns.catplot(x='n_ingredients', y='ingredients', kind='bar', data=most_used, color='#2b8cbe')
sns.despine(bottom = True, left = True) #removes frames
plot.fig.tick_params(left=False)

plt.title('Top 10 ingredients in the Indian food')
plt.xlabel('number of times')
plt.ylabel('ingredients')

plt.show()
```
