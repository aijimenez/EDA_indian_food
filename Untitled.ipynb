```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
pd.set_option('display.max_rows', 426)
indian_food = pd.read_csv("indian_food.csv")
```

```python
indian_food.shape
```

```python
indian_food['num_ingr'] = indian_food['ingredients'].apply(lambda x: len(x.split(',')))
indian_food.head(2)
```

```python
plt.figure()
sns.histplot(indian_food, x="num_ingr", bins=9)
plt.show()
```

```python
# use the function regplot to make a scatterplot
sns.regplot(x=indian_food["num_ingr"], y=indian_food["prep_time"])
plt.show()
```

```python
no_720 = (indian_food['cook_time'] == 720) | (indian_food['cook_time'] == -1) 
food = indian_food.loc[~no_720]
# use the function regplot to make a scatterplot
sns.regplot(x=food["num_ingr"], y=food["cook_time"])
plt.show()
```

```python
# Use the 'hue' argument to provide a factor variable
sns.lmplot( x="cook_time", y="prep_time", data=food, fit_reg=False, hue='region', legend=True)
plt.show()
```

```python
# Custom the histogram:
sns.jointplot(x=food["num_ingr"], y=food["prep_time"], kind='hex', marginal_kws=dict(bins=30, fill=True))

plt.show()
```
